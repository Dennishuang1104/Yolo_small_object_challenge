```python
import pandas as pd

# 讀入資料
df1 = pd.read_csv("image_analysis_report.csv")  # 第一份資料
df2 = pd.read_csv("detection_test.csv")  # 第二份資料

merged = pd.merge(df1, df2, left_on="filename", right_on="name", how="inner")

# 計算每張圖的面積
merged["area"] = merged["width"] * merged["height"]

# 使用 qcut 將圖像分成三類：小（0）、中（1）、大（2）
merged["size_category"] = pd.qcut(merged["area"], q=3, labels=[0, 1, 2], duplicates='drop')

# 計算差異與比例
merged["sahi_diff"] = merged["sahi_count"] - merged["origin_model_count"]
merged["merge_diff"] = merged["merge_count"] - merged["origin_model_count"]
merged["sahi_ratio"] = merged["sahi_diff"] / merged["origin_model_count"]

# 按照比例排序
result = merged.sort_values("sahi_ratio", ascending=False)

# 顯示部分結果
result[["filename", "brightness", "contrast", "origin_model_count", "sahi_count", "sahi_ratio", "size_category"]]
```

```python
df = pd.read_csv('yolov7_pytorch_modules/detection_comparison_final.csv')
plt.figure(figsize=(12, 6))
df[['origin_model_count', 'sahi_count', 'merge_count']].plot(kind='box')
plt.title('Box Plot of Counts: origin_model_count, sahi_count, merge_count')
plt.ylabel('Count')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df["diff"] = df["origin_model_count"] - df["sahi_count"]

# 排序，origin 高在左，sahi 高在右
df_sorted = df.sort_values("diff", ascending=False)

# 顏色標示
colors = df_sorted["diff"].apply(lambda x: "green" if x > 0 else "red")

# 繪圖
plt.figure(figsize=(18, 6))
plt.bar(df_sorted["name"], df_sorted["diff"], color=colors)

# 美化圖表
plt.xticks(rotation=90, fontsize=6)
plt.ylabel("origin_model_count - sahi_count")
plt.title("Difference between Origin and SAHI Counts")
plt.axhline(0, color='gray', linewidth=0.8, linestyle='--')

plt.tight_layout()
plt.show()
```

```python
# 計算差異
df["diff"] = df["origin_model_count"] - df["sahi_count"]

# 取得差異最大的前後 10 筆資料
top_n = 10
df_top = df.nlargest(top_n, "diff")
df_bottom = df.nsmallest(top_n, "diff")
df_filtered = pd.concat([df_top, df_bottom])

# 根據差異排序（左邊 origin 比較高，右邊 SAHI 比較高）
df_filtered = df_filtered.sort_values("diff", ascending=False)

# 設定顏色（綠色表示 origin 高，紅色表示 sahi 高）
colors = df_filtered["diff"].apply(lambda x: "green" if x > 0 else "red")
labels = df_filtered["name"]

# 畫圖
plt.figure(figsize=(14, 6))
bars = plt.bar(labels, df_filtered["diff"], color=colors)

# 美化圖表
plt.xticks(rotation=75, ha='right', fontsize=8)
plt.ylabel("Difference (origin_model_count - sahi_count)", fontsize=10)
plt.title(f"Top {top_n} Origin Higher (Green) vs SAHI Higher (Red) Differences", fontsize=13, fontweight='bold')
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

# 加入 legend（圖例）
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='green', label='Origin > SAHI'),
    Patch(facecolor='red', label='SAHI > Origin')
]
plt.legend(handles=legend_elements, loc='upper left')

# 顯示圖表
plt.tight_layout()
plt.show()
```
